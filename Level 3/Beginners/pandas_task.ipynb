```python
import pandas as pd
import numpy as np

url = 'https://drive.google.com/file/d/1HDgNF7wo3SxZ1Lh-nap8y3LD_nJ0g6A7/view'
file_id = url.split('/')[-2]

dwn_url = 'https://drive.google.com/uc?id=' + file_id
df = pd.read_csv(dwn_url, index_col=0)
df.head()
```

```python
# 1. print first and last five rows
pd.concat([df.head(), df.tail()])
```

```python
# 2. Replace all column values which contain ‘?’ and n.a with NaN.
df.replace(np.nan, 0, inplace=True)
df.loc[df['price']==0]['price']
```

```python
# 3. Print the most expensive car’s company name and price.
df[df['price'] == df['price'].max()][['price', 'company']]
```

```python
# 4. Print All Toyota Cars details
df[(df['company'] == 'toyota')]
```

```python
# 5. Count total cars per company
# df_total = df[['company']].value_counts()
df_total = df.groupby(df['company'])['company'].agg(total_cars = 'count')
df_total
```

```python
# 6. Find each company’s Highest price car
df.groupby(df['company'])['price'].agg(highest_price = max)
```

```python
# 7. Find the average mileage of each car making company
df.groupby(df['company'])['average-mileage'].agg(average_mileage='mean')
```

```python
# 8. Sort all cars by Price column
df.sort_values(by='price')
```

```python
# 9. concat
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}

GermanCars  = pd.DataFrame(GermanCars)
japaneseCars =  pd.DataFrame(japaneseCars)
cars = pd.concat([GermanCars, japaneseCars], keys=['GermanCars', 'japaneseCars'], names=['Key', 'ID'])
cars
```

```python
# 10. merge
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}

Car_Price = pd.DataFrame(Car_Price)
car_Horsepower = pd.DataFrame(car_Horsepower)
data = Car_Price.merge(car_Horsepower, right_on='Company', left_on='Company')
data
```
